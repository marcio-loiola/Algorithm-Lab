## ‚ùå Nega√ß√£o (NOT) ‚Äî ‚ÄúN√ÉO‚Äù

O operador **NOT** funciona como um ‚ÄúN√ÉO‚Äù mesmo.

### üò¥ Problem√°tica
Voc√™ s√≥ pode dormir at√© tarde **se N√ÉO tiver aula**.

Ou seja:
- `tenho_aula = True` ‚Üí n√£o posso dormir at√© tarde  
- `tenho_aula = False` ‚Üí posso dormir tranquilamente üò¥

### üìå Proposi√ß√£o
P: Tenho aula  
¬¨P ‚Üí N√ÉO tenho aula


```python
tenho_aula = True  # Amanh√£ tem aula :(

if not tenho_aula:
    print("üò¥ Posso dormir at√© tarde!")
else:
    print("‚è∞ N√£o posso dormir at√© tarde... tenho aula.")
```

```python
tenho_aula = False   # hoje √© s√≥ alegria üòé

if not tenho_aula:
    print("üò¥ Posso dormir at√© tarde! Hoje √© dia de descanso!")
else:
    print("‚è∞ Tenho aula... acordar cedo √© obrigat√≥rio.")
```

### üç∞ Problem√°tica
Voc√™ quer fazer aquele bolo top pra postar no Instagram.
